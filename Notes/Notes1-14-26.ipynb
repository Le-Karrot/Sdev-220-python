Create a variable for string or number, since as practice you shouldn't really place the number in there. Since you might to change it later, and the value will need to be change in all instances and not just once in a variable. Causeing more work and inefficiency.

```python
## not too practical

hours_in_day: int = 24

hours_in_week = hours_in_day * 5

## since we might need to change 5 and if we have 5 in all instances for this calculation we'll have to find and change all of them
## rather do this

hours_in_day: int = 24
days_in_week: int = 5

hours_in_week = hours_in_day * days_in_week

print(hours_in_week)
```
